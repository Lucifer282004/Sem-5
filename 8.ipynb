```python
def binary(arr,key,l,r):
  while(l<=r):
    mid=(l+r)//2

    if(arr[mid]==key):
      return mid
    elif (arr[mid]<key):
      return binary (arr,key,mid+1,r)
    else:
      return binary(arr,key,l,mid-1)
  return -1
if __name__=="__main__":
  arr=[1,2,3,4,5,6,7,8,9]

  print(binary(arr,5,0,len(arr)-1))
```

```python
def mergeSort(arr, l, mid, r):
    n1 = mid - l + 1
    n2 = r - mid

    a = [0] * n1
    b = [0] * n2

    for i in range(n1):
        a[i] = arr[i + l]

    for j in range(n2):
        b[j] = arr[mid + 1 + j]


    i = 0;j = 0;k = l

    while i < n1 and j < n2:
        if a[i] < b[j]:
            arr[k] = a[i]
            i += 1
        else:
            arr[k] = b[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = a[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = b[j]
        j += 1
        k += 1


def merge(arr, l, r):
    if l < r:
        mid = (l + r) // 2

        merge(arr, l, mid)
        merge(arr, mid + 1, r)
        mergeSort(arr, l, mid, r)

if __name__ == "__main__":
    arr = [5, 4, 3, 2, 1]
    print("Original array:", arr)

    merge(arr, 0, len(arr) - 1)

    print("Sorted array:", arr)
```

```python

def insertionSort(arr, left, right):
    for i in range(left + 1, right + 1):
        key = arr[i]
        j = i - 1
        while j >= left and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key


def merge(arr, l, m, r):
    len1, len2 = m - l + 1, r - m
    left, right = [], []

    for i in range(0, len1):
        left.append(arr[l + i])
    for i in range(0, len2):
        right.append(arr[m + 1 + i])

    i, j, k = 0, 0, l

    while i < len1 and j < len2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    while i < len1:
        arr[k] = left[i]
        i += 1
        k += 1

    while j < len2:
        arr[k] = right[j]
        j += 1
        k += 1


def timSort(arr):
    RUN = 32

    n = len(arr)


    for i in range(0, n, RUN):
        insertionSort(arr, i, min((i + RUN - 1), n - 1))


    size = RUN
    while size < n:
        for left in range(0, n, 2 * size):
            mid = min(n - 1, left + size - 1)
            right = min((left + 2 * size - 1), (n - 1))

            if mid < right:
                merge(arr, left, mid, right)

        size = 2 * size


if __name__ == "__main__":
    arr = [3,2,7,9,4,1,5,6,8]
    print("Original array:", arr)

    timSort(arr)

    print("Sorted array using TimSort:", arr)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None


    def add(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            last = self.head
            while last.next:
                last = last.next
            last.next = new_node


    def print_list(self):
        curr = self.head
        while curr:
            print(f"{curr.data} -> ", end="")
            curr = curr.next
        print("NULL")

    # i) Find the size of the linked list
    def size(self):
        count = 0
        curr = self.head
        while curr:
            count += 1
            curr = curr.next
        return count

    # ii) Search for a specific item
    def search(self, item):
        curr = self.head
        while curr:
            if curr.data == item:
                return True
            curr = curr.next
        return False

    # iii) Access a specific item using index value
    def get(self, index):
        curr = self.head
        count = 0
        while curr:
            if count == index:
                return curr.data
            count += 1
            curr = curr.next
        return None

    # iv) Set a new value of an item using index value
    def set(self, index, data):
        curr = self.head
        count = 0
        while curr:
            if count == index:
                curr.data = data
                return
            count += 1
            curr = curr.next
        print("Index out of range")

    # v) Delete the first item from the linked list
    def delete_first(self):
        if self.head is None:
            return
        self.head = self.head.next

    # vi) Delete the last item from the linked list
    def delete_last(self):
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
            return
        second_last = self.head
        while second_last.next.next:
            second_last = second_last.next
        second_last.next = None


if __name__ == "__main__":
    l = LinkedList()

    l.add(1)
    l.add(2)
    l.add(3)

    l.print_list()

    # i) Find size of the list
    print("Size of the linked list:", l.size())

    # ii) Search for a specific item
    print("Is 2 in the list?", l.search(2))
    print("Is 5 in the list?", l.search(5))

    # iii) Access specific item by index
    print("Item at index 1:", l.get(1))
    print("Item at index 3:", l.get(3))


    l.set(1, 10)
    l.print_list()

    # v) Delete the first item
    l.delete_first()
    l.print_list()

    # vi) Delete the last item

    l.delete_last()
    l.print_list()
```
